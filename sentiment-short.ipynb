```python
import pandas as pd
import os

# 🔹 Direct file path (recommended)
path = "/kaggle/input/competitions/gen-ai-summarization-sprint/samsum-test.csv"  
df = pd.read_csv(path)

# Logic: Search for happy words
def get_sentiment(text):
    text = str(text).lower()
    happy_words = ['good', 'great', 'amazing', 'excellent', 'happy', 'best', 'love']
    
    for word in happy_words:
        if word in text:
            return "Positive"
    
    return "Negative"

# Apply sentiment
df['sentiment'] = df['dialogue'].apply(get_sentiment)

# Save output
df[['id', 'sentiment']].to_csv('submission.csv', index=False)

print("Done!")
```
