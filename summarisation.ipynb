```python
import pandas as pd
import re

# Direct file path (already a file)
file_path = "/kaggle/input/competitions/gen-ai-summarization-sprint/samsum-test.csv"

df = pd.read_csv(file_path)

def simple_clean(text):
    clean = re.sub(r'[^a-zA-Z\s]', '', str(text))
    words = clean.split()
    return " ".join(words[:10]).capitalize() + "."

# Use second column safely
df['summary'] = df.iloc[:, 1].apply(simple_clean)

df.to_csv('submission.csv', index=False)

print("Done successfully!")
```
